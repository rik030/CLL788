```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
  

data_1 = pd.read_excel("CLL788_A1/data_1.xlsx")

data_1.head()   
```

```python
X=data_1['x'].to_numpy()
Y=data_1['y'].to_numpy()
```

```python
#Scatter Plot
plt.scatter(X,Y)
plt.show()
```

```python
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 6))
ax.hist(X)
 
# Show plot
plt.show()
```

```python
# Creating histogram
fig, ay = plt.subplots(figsize =(10, 6))
ay.hist(Y)
 
# Show plot
plt.show()
```

```python
#HeatMap
import seaborn as sns
sns.heatmap(data_1.corr(), annot = True, fmt='.2g',cmap= 'coolwarm')
```

```python
#Boxplot for X
fig = plt.figure(figsize =(10, 7))
plt.boxplot(X)
plt.show()
```

```python
#Boxplot for Y
fig = plt.figure(figsize =(10, 7))
plt.boxplot(Y)
plt.show()
```

```python
import statistics as st
mean_x=np.mean(X)
median_x=np.median(X)
mode_x=st.mode(X)
std_x=np.std(X)
print("Mean of X: ", mean_x)
print("Median of X: ", median_x)
print("Mode of X: ", mode_x)
print("SD of X: ", std_x)
```

```python
import statistics as st
mean_y=np.mean(Y)
median_y=np.median(Y)
mode_y=st.mode(Y)
std_y=np.std(Y)
print("Mean of Y: ", mean_y)
print("Median of Y: ", median_y)
print("Mode of Y: ", mode_y)
print("SD of Y: ", std_y)
```

```python
import statistics as st
mean_x=np.mean(X)
median_x=np.median(X)
mode_x=st.mode(X)
std_x=np.std(X)
print("Mean of X: ", mean_x)
print("Median of X: ", median_x)
print("Mode of X: ", mode_x)
print("SD of X: ", std_x)
```

```python
import statistics as st
mean_y=np.mean(Y)
median_y=np.median(Y)
mode_y=st.mode(Y)
std_y=np.std(Y)
print("Mean of Y: ", mean_y)
print("Median of Y: ", median_y)
print("Mode of Y: ", mode_y)
print("SD of Y: ", std_y)
```

```python
#outlier using standard deviation in X
Outlier_X=[]
k=3
for i in range(0, len(X)):
    if((abs(X[i]-mean_x)/std_x)>k):
        Outlier_X.append(X[i])
print(Outlier_X)        
```

```python
#outlier using standard deviation in Y
Outlier_Y=[]
k=3
for i in range(0, len(Y)):
    if((abs(Y[i]-mean_y)/std_y)>k):
        Outlier_Y.append(Y[i])
print(Outlier_Y)  
```

```python
#Outlier detection : Median Absolute Deviation in X
Outlier_X=[]
k=3
MAD_x=np.median(abs(X-median_x))
for i in range(0, len(X)):
    if((0.6745*(X[i]-median_x))/MAD_x>k):
        Outlier_X.append(X[i])
print(Outlier_X)
```

```python
#Outlier detection : Median Absolute Deviation in Y
Outlier_Y=[]
k=3
MAD_y=np.median(abs(Y-median_y))
for i in range(0, len(Y)):
    if((0.6745*(Y[i]-median_y))/MAD_y>k):
        Outlier_Y.append(Y[i])
print(Outlier_Y)
```
